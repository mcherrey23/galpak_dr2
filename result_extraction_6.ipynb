**The only difference with the previous version is that the constrains in Dv are released from 1000kms to 500 kms.**

The result extraction is automated. The result table only contains the best run.

In this version, we still distinguish primary galaxies from isolated.
- Primary galaxies are the closest of the LOS (when multiple are present), with a closest neighbour at least at b_sep (not high sthg in 10 - 30 kpc). 

- Isolated galaxies are chosen conservatively: we only conside galaxies alone in 100 kpc and +-1000kms (no concept of satellite galaxies with low mass) and no neighbour closer than B +b_sep (to avoid cases like 2 galaxies at resp 97 and 103 kpc).
